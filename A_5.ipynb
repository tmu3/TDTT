```python
#W5A1
def max_of_two(a, b):
  return max(a, b)

print (max_of_two(23, 36))
```

```python
#W5A2
def swap(a, b):
  return b, a

print (swap(23, 36))
```

```python
#W5A3
def snt(n):
  if n < 1:
    return False
  elif n == 2:
    return True
  else:
    for i in range (2, n + 1):
      if n % i == 0:
        return False
      else:
        return True

snt(11)
```

```python
#W5A4
def shh(n):
  sum = 0
  for i in range (1, n):
    if n % i == 0:
      sum += i
  if sum == n:
    return True
  else:
    return False

shh(28)
```

```python
#W5A5
def duyet_list(n, k):
  for i in n:
    if i == k:
      return n.index(i)
  return -1

duyet_list([1, 2, 3, 4, 5], 5)
```

```python
#W5A6
def fac(n):
  fac = 1
  for i in range (1, n+1):
    fac *= i
  return fac

print (fac(5))
```

```python
#W5A7
def calculator(a, b, operator):
  if operator == '+':
    result = a + b
  elif operator == '-':
    result = a - b
  elif operator == '*':
    result = a * b
  elif operator == '/':
    if b == 0:
      return "Lỗi: không thể chia cho 0"
      result = a / b
  else:
    return "Lỗi: phép toán không hợp lệ"

  return round(result, 2)

print (calculator(2, 3, '+'))
```

```python
#W5A8
def kc_hamming(x, y):
  xor = x ^ y
  return bin(xor).count('1')

kc_hamming(2, 3)
```

```python
#W5A9
def sum_digits(n):
  total = 0
  while n > 0:
    total += n % 10
    n //= 10
  return total

sum_digits(123)
```

```python
#W5A10
def dang_cau(str1, str2):
  dic1 = {}
  dic2 = {}
  for char1, char2 in zip(str1, str2):
    if char1 in dic1:
      if dic1[char1] != char2:
        return False
    else:
      dic1[char1] = char2
    if char2 in dic2:
      if dic2[char2] != char1:
        return False
    else:
      dic2[char2] = char1
  return True

dang_cau('egg', 'add')
```
