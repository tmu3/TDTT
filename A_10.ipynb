```python
#W10A3
def is_happy(n):
  seen = set()
  while n != 1:
    if n in seen:
      return False
    seen.add(n)
    n = sum(int(d)**2 for d in str(n))
  return True

n = int(input("Nhập số n: "))
try:
  if is_happy(n):
    print("YES")
  else:
    print("NO")
except:
  print("ValueError thêm set")
```

```python
#W10A10
n = int(input("Nhập số lượng phần tử: "))
lst = []
for i in range(n):
  value = int(input("Nhập phần tử: "))
  lst.append(value)
seen = []
total = 0
try:
  for i in lst:
    if i in seen:
      raise ValueError
    else:
      total += i
      seen.append(i)
except ValueError:
  print("Mang khong hop le")
```
