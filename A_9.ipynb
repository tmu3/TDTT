```python
#W9A1
def merge(a, b):
  if not a:
    return b
  if not b:
    return a
  if a[0] < b[0]:
    return [a[0]] + merge(a[1:], b)
  return [b[0]] + merge(a, b[1:])

a = list(map(int, input("Nhập dãy 1: ").split()))
b = list(map(int, input("Nhập dãy 2: ").split()))
print(*merge(a, b))
```

```python
#W9A2
def count_le(matrix, x):
  n, m = len(matrix), len(matrix[0])
  i, j = 0, m - 1
  count = 0
  while i < n and j >= 0:
    if matrix[i][j] <= x:
      count += (j + 1)
      i += 1
    else:
      j -= 1
  return count

def kth_smallest(matrix, k):
  n, m = len(matrix), len(matrix[0])
  def binary_search(low, high):
    if low > high:
      return low
    mid = (low + high) // 2
    cnt = count_le(matrix, mid)
    if cnt >= k:
      return binary_search(low, mid - 1)
    else:
      return binary_search(mid + 1, high)
  return binary_search(matrix[0][0], matrix[n-1][m-1])

n, m, k = map(int, input("Nhập n, m, k: ").split())
matrix = []
for i in range(n):
  matrix.append(list(map(int, input(f"Nhập dòng {i + 1}: ").split())))
print(kth_smallest(matrix, k - 1))
```

```python
#W9A3
def min_swap(num):
  s = list(str(num))

  def min_num(i):
    if i == len(s):
      return
    min_digit = min(s[i+1:], default=s[i])
    if min_digit < s[i]:
      j = len(s) - 1
      while s[j] != min_digit:
        j -= 1
      s[i], s[j] = s[j], s[i]
      return
    else:
      min_num(i + 1)

  min_num(0)
  return int("".join(s))

n = int(input("Nhập số n: "))
print(min_swap(n))
```

```python
#W9A4
def can_transform(a, b):
  n, m = len(a), len(b)
  def dfs(i, j):
    if j == m:
      return all(c.islower() for c in a[i:])
    if i == n:
      return False

    if a[i].islower():
      if dfs(i + 1, j):
          return True
      if a[i].upper() == b[j] and dfs(i + 1, j + 1):
          return True
      return False

    else:
      if a[i] == b[j]:
        return dfs(i + 1, j + 1)
      return False

  return dfs(0, 0)

q = int(input("Nhập số lượng câu hỏi: "))
for i in range(q):
  a = input("Nhập câu a: ")
  b = input("Nhập câu b: ")
  if can_transform(a, b):
    print("YES")
  else:
    print("NO")
```

```python
#W9A5
def in_list(a, b):
  def match(i, j):
    if i == len(a):
      return True
    if j == len(b):
      return False
    if a[i] == b[j]:
      return match(i + 1, j + 1)
    return False

  def check(j):
    if j == len(b):
      return False
    if match(0, j):
      return True
    return check(j + 1)

  return check(0)

a = list(map(int, input("Nhập dãy a: ").split()))
b = list(map(int, input("Nhập dãy b: ").split()))
if in_list(a, b):
  print("YES")
else:
  print("NO")
```

```python
#W9A6
def hvtt(a):
  n = len(a)
  i = n - 2
  while i >= 0 and a[i] >= a[i + 1]:
    i -= 1
  if i < 0:
    a.reverse()
    return a

  j = n - 1
  while a[j] <= a[i]:
    j -= 1

  a[i], a[j] = a[j], a[i]

  a[i + 1:] = reversed(a[i + 1:])

  return a

n = int(input("Nhập số lượng phần tử: "))
a = list(map(int, input("Nhập dãy a: ").split()))
print(*hvtt(a))
```

```python
#W9A7
def canPlaceFlowers(flowerbed, n):
  length = len(flowerbed)
  def dfs(i, n):
    if n <= 0:
      return True
    if i >= length:
      return False

    if flowerbed[i] == 0:
      left_empty = (i == 0 or flowerbed[i - 1] == 0)
      right_empty = (i == length - 1 or flowerbed[i + 1] == 0)
      if left_empty and right_empty:
          flowerbed[i] = 1
          if dfs(i + 2, n - 1):
            return True
          flowerbed[i] = 0

    return dfs(i + 1, n)

  return dfs(0, n)

garden = list(map(int, input("Nhập dãy garden: ").split()))
n = int(input("Nhập số lượng cần đặt: "))
print(canPlaceFlowers(garden, n))
```

```python
#W9A8
def is_path_crossing(path):
  visit = {(0, 0)}
  def dfs(i, x, y):
    if i == len(path):
      return False
    if path[i] == 'U':
        y += 1
    elif path[i] == 'D':
        y -= 1
    elif path[i] == 'L':
        x -= 1
    elif path[i] == 'R':
        x += 1
    if (x, y) in visit:
        return True
    visit.add((x, y))
    return dfs(i + 1, x, y)

  return dfs(0, 0, 0)

path = input("Nhập đường đi: ")
print(is_path_crossing(path))
```

```python
#W9A9
def median(a, b, c, d, e):
  if a > b:
    return median(b, a, c, d, e)

  if c > d:
    return median(a, b, d, c, e)

  if a > c:
    return median(c, d, a, b, e)

  if b > d:
    return median(a, d, c, b, e)

  if e < b:
    return b
  if e > c:
    return c
  return e

a, b, c, d, e = map(int, input("Nhập 5 số: ").split())
print(median(a, b, c, d, e))
```

```python
#W9A10
def min_dist(h, heaters, i=0):
  if i == len(heaters):
    return float('inf')
  return min(
      abs(h - heaters[i]),
      min_dist(h, heaters, i + 1)
  )

def find_radius(houses, heaters, i=0):
  if i == len(houses):
    return 0
  return max(
    min_dist(houses[i], heaters),
    find_radius(houses, heaters, i + 1)
  )

houses = list(map(int, input("Nhập vị trí nhà: ").split()))
heaters = list(map(int, input("Nhập vị trí lò sưởi: ").split()))
print(find_radius(houses, heaters))
```
