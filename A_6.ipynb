```python
#W6A17
n = int(input("Nhập số hàng và cột: "))
matrix = []
for i in range(n):
  row = list(map(int, input(f"Hàng {i+1}: ").split()))
  matrix.append(row)
main_diag = [matrix[i][i] for i in range(n)]
anti_diag = [matrix[i][n-1-i] for i in range(n)]
print (main_diag)
print (anti_diag)
```

```python
#W6A18
n, m, k = map(int, input("Nhập số hàng, cột và k: ").split())
matrix = []
for i in range(n):
  row = list(map(int, input(f"Hàng {i+1}: ").split()))
  matrix.append(row)
sum_row = 0
sum_col = 0
for i in range (m):
  sum_row += matrix[k][i]
for i in range (n):
  sum_col += matrix[i][k]
print (f"Tổng hàng {k+1}: {sum_row}")
print (f"Tổng cột {k+1}: {sum_col}")
```

```python
#W6A21
def position(n):
  dic = {}
  for idx, val in enumerate(n):
    if val not in dic:
      dic[val] = []
    dic[val].append(idx)
  for key in dic:
    dic[key] = tuple(dic[key])
  return dic

lst = input("Nhập dãy: ").split()
print (position(lst))
```
