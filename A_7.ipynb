```python
#W7A1
def binary_search(arr, k):
  left = 0
  right = len(arr) - 1
  while left <= right:
    mid = (left + right) // 2
    if arr[mid] == k:
      return mid
    elif arr[mid] < k:
      left = mid + 1
    else:
      right = mid - 1
  return -1

a = list(map(int, input("Nhập dãy số nguyên tăng dần: ").split()))
k = int(input("Nhập số nguyên: "))
print(binary_search(a, k))
```

```python
#W7A3
def bubble_sort(arr):
  n = len(arr)
  for i in range(n):
    for j in range(0, n - i - 1):
      if arr[j] > arr[j + 1]:
        arr[j], arr[j + 1] = arr[j + 1], arr[j]
  return arr

def selection_sort(arr):
  n = len(arr)
  for i in range(n):
    min_index = i
    for j in range(i + 1, n):
      if arr[j] < arr[min_index]:
        min_index = j
    arr[i], arr[min_index] = arr[min_index], arr[i]
  return arr

a = list(map(int, input("Nhập dãy số nguyên: ").split()))
for i in bubble_sort(a) and selection_sort(a):
  print(i, end = " ")
```

```python
#W7A4
def count_occurances(arr):
  max_count = 0
  result = arr[0]
  for i in range(len(arr)):
    count = 0
    for j in range(len(arr)):
      if arr[j] == arr[i]:
        count += 1
    if count > max_count:
      max_count = count
      result = arr[i]
  return result, max_count

a = list(map(int, input("Nhập dãy số nguyên: ").split()))
print(count_occurances(a)[0], "xuất hiện nhiều nhất, sớm nhất,",count_occurances(a)[1], "lần")
```

```python
#W7A5
a = list(map(int, input("Nhập dãy số nguyên: ").split()))
k = int(input("Nhập số nguyên: "))
count = 0
seen = set()
for i in range (0, len(a) - 1):
  for j in range (i + 1, len(a) - 1):
    if (a[i] + a[j]) % k == 0:
      pair = tuple(sorted((a[i], a[j])))
      if pair not in seen:
        count += 1
        seen.add(pair)
print(count)
```

```python
#W7A7
def list_length(a):
  dp = [1] * len(a)
  for i in range(len(a)):
    for j in range(i):
      if a[j] < a[i]:
        dp[i] = max(dp[i], dp[j] + 1)
  return max(dp)

n = eval(input("Nhập list: "))
print(list_length(n))
```

```python
#w7A8
interval = list(eval("[" + input("Nhập các khoảng thời gian: ") + "]"))
queries = list(eval(input("Nhập các truy vấn: ")))
interval.sort(key=lambda x: x[1] - x[0] + 1)
result = []
for q in queries:
  ans = -1
  for l, r in interval:
    if l <= q <= r:
      ans = r - l + 1
      break
  result.append(ans)
print(result)
```

```python
#W7A13
def dhh(x):
  freq = {}
  for i in x:
    if i in freq:
      freq[i] += 1
    else:
      freq[i] = 1
  max_len = 0
  for i in freq:
    if i + 1 in freq:
      max_len = max(max_len, freq[i] + freq[i + 1])
  return max_len

arr = list(map(int, input("Nhập dãy số nguyên: ").split()))
print(dhh(arr))
```
