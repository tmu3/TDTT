```python
#W8A1
class Cylinder:
  def __init__(self, radius, height):
    self.radius = radius
    self.height = height

  def area(self):
    return 2 * 3.14 * self.radius * (self.radius + self.height)

  def volume(self):
    return 3.14 * self.radius ** 2 * self.height

r, h = map(int, input("Nhập bán kính đáy, chiều cao: ").split())
c = Cylinder(r, h)
print(f"Diện tích bề mặt: {round(c.area(), 2)}")
print(f"Thể tích: {round(c.volume(), 2)}")
```

```python
#W8A2
class Date:
  def __init__(self, day, month, year):
    self.day = day
    self.month = month
    self.year = year

  def check_leap_year(self):
    if self.year % 400 == 0:
      return True
    elif self.year % 100 == 0:
      return False
    elif self.year % 4 == 0:
      return True
    else:
      return False

  def check_valid(self):
    if self.year < 0:
      return False
    if self.month < 1 or self.month > 12:
      return False
    if self.day < 1 or self.day > 31:
      return False
    if self.month == 2:
      if self.check_leap_year():
        if self.day > 29:
          return False
      else:
        if self.day > 28:
          return False
    elif self.month == 4 or self.month == 6 or self.month == 9 or self.month == 11:
      if self.day > 30:
        return False
    elif self.month == 1 or self.month == 3 or self.month == 5 or self.month == 7 or self.month == 8 or self.month == 10 or self.month == 12:
      if self.day > 31:
        return False
    return True

  def result(self):
    if self.month == 12 and self.day == 31:
      self.day = 1
      self.month = 1
      self.year += 1

    elif self.month == 4 or self.month == 6 or self.month == 9 or self.month == 11:
      if self.day == 30:
        self.day = 1
        self.month += 1
      else:
        self.day += 1

    elif self.month == 2:
      if self.check_leap_year():
        if self.day == 29:
          self.day = 1
          self.month += 1
        else:
          self.day += 1
      else:
        if self.day == 28:
          self.day = 1
          self.month += 1
        else:
          self.day += 1

    else:
      if self.day == 31:
        self.day = 1
        self.month += 1
      else:
        self.day += 1

d, m, y = map(int, input("Nhập ngày, tháng, năm: ").split("/"))
x = Date(d, m, y)
x.result()
if not x.check_valid():
   print("INVALID")
else:
  print(f"Ngày mai: {x.day}/{x.month}/{x.year}")
```

```python
#W8A3
class Point:
  def __init__(self, x, y):
    self.x = x
    self.y = y

  def check(self):
    horizontal_axes = False
    vertical_axes = False
    if self.x == 0 and self.y == 0:
      horizontal_axes = True
      vertical_axes = True
      distance = 0
    elif self.x == 0:
      vertical_axes = True
      distance = abs(self.y)
    elif self.y == 0:
      horizontal_axes = True
      distance = abs(self.x)
    else:
      distance = round((self.x ** 2 + self.y ** 2)**(1/2), 2)
    return horizontal_axes, vertical_axes, distance

x, y = map(int, input("Nhập tọa độ điểm: ").split())
n = Point(x, y)
horizontal_axes, vertical_axes, distance = n.check()
if horizontal_axes == True and vertical_axes == True:
  print("M là gốc toạ độ.")
elif vertical_axes == True:
  print("M trên trục tung.")
elif horizontal_axes == True:
  print("M trên trục hoành")
print(f"Khoảng cách M đến gốc tọa độ: {distance}")
```

```python
#W8A4
class Calculator:
  def __init__(self, n1, n2):
    self.n1 = n1
    self.n2 = n2

  def add(self):
    return self.n1 + self.n2

  def subtract(self):
    return self.n1 - self.n2

  def product(self):
    return self.n1 * self.n2

  def divide(self):
    return self.n1 / self.n2

  def power(self):
    return self.n1 ** self.n2

  def mod(self):
    return self.n1 % self.n2

  def set_number(self):
    self.n1 = int(input("Nhập số thứ nhất: "))
    self.n2 = int(input("Nhập số thứ hai: "))

n1, n2 = map(int, input("Nhập 2 số: ").split())
c = Calculator(n1, n2)
while True:
  print("\n")
  print("1. Cộng")
  print("2. Trừ")
  print("3. Nhân")
  print("4. Chia")
  print("5. Lũy thừa")
  print("6. Chia dư")
  print("7. Nhập lại 2 số")
  print("0. Thoát")

  choice = input("Chọn chức năng: ")

  if choice == "1":
    print("Kết quả:", c.add())
  elif choice == "2":
    print("Kết quả:", c.subtract())
  elif choice == "3":
    print("Kết quả:", c.product())
  elif choice == "4":
    print("Kết quả:", c.divide())
  elif choice == "5":
    print("Kết quả:", c.power())
  elif choice == "6":
    print("Kết quả:", c.mod())
  elif choice == "7":
    c.set_number()
  elif choice == "0":
    stop = input("Bạn có muốn thoát: ").lower()
    if stop == "yes":
      break
```

```python
#W8A5
class ShoppingCart:
  def __init__(self):
    self.cart = []

  def add_product(self, ten, gia):
    product = {"ten": ten, "gia": gia}
    self.cart.append(product)

  def remove_product(self, ten):
    for product in self.cart:
      if product["ten"] == ten:
        self.cart.remove(product)
        break

  def is_empty(self):
    return len(self.cart) == 0

  def total_price(self):
    total = 0
    for product in self.cart:
      total += product["gia"]
    return total

  def show_cart(self):
    if self.is_empty():
      print("Giỏ hàng đang rỗng")
    else:
      print("Danh sách sản phẩm trong giỏ:")
      for product in self.cart:
        print(f"- {product['ten']}: {product['gia']}")

  def clear_cart(self):
    self.cart.clear()

cart = ShoppingCart()

while True:
  print("\n")
  print("2. Thêm sản phẩm vào giỏ")
  print("3. Xóa sản phẩm theo tên")
  print("4. Kiểm tra giỏ hàng có rỗng không")
  print("5. Tính tổng tiền giỏ hàng")
  print("6. Hiển thị giỏ hàng")
  print("7. Xóa toàn bộ giỏ hàng")
  print("0. Thoát")

  choice = input("Chọn chức năng: ")

  if choice == "2":
    ten = input("Nhập tên sản phẩm: ")
    gia = int(input("Nhập giá sản phẩm: "))
    cart.add_product(ten, gia)
    print("Đã thêm sản phẩm")

  elif choice == "3":
    ten = input("Nhập tên sản phẩm cần xóa: ")
    if cart.remove_product(ten):
      print("Đã xóa sản phẩm")
    else:
      print("Không tìm thấy sản phẩm")

  elif choice == "4":
    if cart.is_empty():
      print("Giỏ hàng đang rỗng")
    else:
      print("Giỏ hàng không rỗng")

  elif choice == "5":
    print("Tổng tiền giỏ hàng:", cart.total_price())

  elif choice == "6":
    cart.show_cart()

  elif choice == "7":
    cart.clear_cart()
    print("Đã xóa toàn bộ giỏ hàng")

  elif choice == "0":
    stop = input("Có muốn thoát khỏi chương trình không? ").lower()
    if stop == "yes":
      break
```

```python
#W8A6
class Rectangle:
  def __init__(self, width, length):
    self.width = width
    self.length = length

  def area(self):
    return round(self.width * self.length, 2)

  def perimeter(self):
    return round((self.width + self.length)*2, 2)

  def scale(self, k):
    self.width *= k
    self.length *= k

width, length, k = map(int, input("Nhập chiều rộng, chiều dài, k: ").split())
r = Rectangle(width, length)
r.scale(k)
print(f"Diện tích: {r.area()}")
print(f"Chu vi: {r.perimeter()}")
```

```python
#W8A7
class Fraction:
  def __init__(self, num, den):
    self.num = num
    self.den = den

  def rutgon(self):
    ucln = 1
    for i in range(1, min(self.num, self.den) + 1):
      if self.num % i == 0 and self.den % i == 0:
        ucln = i
        new_num = self.num // ucln
        new_den = self.den // ucln
    return Fraction(new_num, new_den)

  def __add__(self, other):
    new_num = self.num * other.den + self.den * other.num
    new_den = self.den * other.den
    return Fraction(new_num, new_den)

  def __sub__(self, other):
    new_num = self.num * other.den - self.den * other.num
    new_den = self.den * other.den
    return Fraction(new_num, new_den)

  def __mul__(self, other):
    new_num = self.num * other.num
    new_den = self.den * other.den
    return Fraction(new_num, new_den)

  def __truediv__(self, other):
    new_num = self.num * other.den
    new_den = self.den * other.num
    return Fraction(new_num, new_den)

  def __str__(self):
    if self.den == 1:
      return str(self.num)
    else:
      return f"{self.num}/{self.den}"

a, b, op, c, d = input("Nhập phép tính: ").split()
a, b, c, d = int(a), int(b), int(c), int(d)
f1 = Fraction(a, b)
f2 = Fraction(c, d)
if op == "+":
  print((f1 + f2).rutgon())
elif op == "-":
  print((f1 - f2).rutgon())
elif op == "*":
  print((f1 * f2).rutgon())
elif op == "/":
  print((f1 / f2).rutgon())
```

```python
#W8A8
class BankAccount:
  def __init__(self, owner, balance):
    self.owner = owner
    self.balance = balance

  def deposit(self, amount):
    self.balance += amount
    return self.balance

  def withdraw(self, amount):
    if amount <= self.balance:
      self.balance -= amount
      return self.balance

o, b = input("Nhập tên chủ tài khoản, số dư: ").split()
b = float(b)
account = BankAccount(o, b)
n = int(input("Nhập số lần giao dịch: "))
for _ in range(n):
  op, amount = input("Nhập giao dịch: ").split()
  op = op.lower()
  amount = float(amount)
  if op == "deposit":
    account.deposit(amount)
  elif op == "withdraw":
    account.withdraw(amount)
print(round(account.balance, 2))
```

```python
#W8A9
class Student:
  def __init__(self):
    self.file = []

  def add(self, subject, score):
    hoc_ba = {"subject": subject, "score": score}
    self.file.append(hoc_ba)

  def avg(self):
    total = 0
    for hoc_ba in self.file:
      total += hoc_ba["score"]
      average = total / len(self.file)
    return round(average, 2)

  def rank(self):
    if self.avg() >= 8:
      return "Excellent"
    elif self.avg() >= 6.5:
      return "Good"
    elif self.avg() >= 5:
      return "Average"
    else:
      return "Poor"

name = input("Nhập tên sinh viên: ")
stu = Student()
n = int(input("Nhập số môn học: "))
for i in range(n):
  sub, sco = input("Nhập môn học và điểm: ").split()
  sco = float(sco)
  stu.add(sub, sco)
print(f"{name} {stu.avg()} {stu.rank()}")
```

```python
#W8A10
class Book:
  def __init__(self, title, author, year):
    self.title = title
    self.author = author
    self.year = year

class Library():
  def __init__(self):
    self.bookshelf = []

  def add(self, title, author, year):
    book = Book(title, author, year)
    self.bookshelf.append(book)

  def count_by_author(self, author):
    count = 0
    for book in self.bookshelf:
      if book.author == author:
        count += 1
    return count

  def count_by_year(self, year):
    count = 0
    for book in self.bookshelf:
      if book.year == year:
        count += 1
    return count

lib = Library()
n = int(input("Nhập số thao tác: "))
for i in range(n):
  line = input().strip()
  if line.startswith("ADD"):
    _, data = line.split()
    title, author, year = data.split(";")
    lib.add(title, author, int(year))
  elif line.startswith("COUNT"):
    _, author = line.split()
    print(lib.count_by_author(author))
  elif line.startswith("COUNTYEAR"):
    _, year = line.split()
    print(lib.count_by_year(int(year)))
```

```python
#W8A11
import math
class Point:
  def __init__(self, x, y, z):
    self.x = x
    self.y = y
    self.z = z

  def vecto(self, other):
    x = self.x - other.x
    y = self.y - other.y
    z = self.z - other.z
    return Point(x, y, z)

  def vecto_pt(self, other):
    x = self.y * other.z - self.z * other.y
    y = self.z * other.x - self.x * other.z
    z = self.x * other.y - self.y * other.x
    return Point(x, y, z)

  def angle(self, other):
    dot_product = self.x * other.x + self.y * other.y + self.z * other.z
    magnitude_self = math.sqrt(self.x**2 + self.y**2 + self.z**2)
    magnitude_other = math.sqrt(other.x**2 + other.y**2 + other.z**2)
    cos_phi = dot_product / (magnitude_self * magnitude_other)
    phi = math.degrees(math.acos(cos_phi))
    return phi

xa, ya, za = map(int, input("Nhập tọa độ điểm A: ").split())
xb, yb, zb = map(int, input("Nhập tọa độ điểm B: ").split())
xc, yc, zc = map(int, input("Nhập tọa độ điểm C: ").split())
xd, yd, zd = map(int, input("Nhập tọa độ điểm D: ").split())
A = Point(xa, ya, za)
B = Point(xb, yb, zb)
C = Point(xc, yc, zc)
D = Point(xd, yd, zd)
AB = B.vecto(A)
AC = C.vecto(A)
BC = C.vecto(B)
BD = D.vecto(B)
ABC = AB.vecto_pt(AC)
BCD = BC.vecto_pt(BD)
phi = ABC.angle(BCD)
print(round(phi, 2))
```

```python
#W8A12
class Complex:
  def __init__(self, real, imag):
    self.real = real
    self.imag = imag

  def mul(self, other):
    real = self.real * other.real - self.imag * other.imag
    imag = self.real * other.imag + self.imag * other.real
    return Complex(real, imag)

r1, i1 = map(int, input("Nhập phần thực và ảo của số phức thứ nhất: ").split())
r2, i2 = map(int, input("Nhập phần thực và ảo của số phức thứ hai: ").split())
c1 = Complex(r1, i1)
c2 = Complex(r2, i2)
c3 = c1.mul(c2)
print(f"{c3.real} {c3.imag}")
```

```python
#W8A13
class Triangle:
  def __init__(self, a, b, c):
    self.a = a
    self.b = b
    self.c = c

  def check_triangle(self):
    if self.a + self.b > self.c and self.a + self.c > self.b and self.b + self.c > self.a:
      return True
    else:
      return False

  def area(self):
    if self.check_triangle() == True:
      p = (self.a + self.b + self.c) / 2
      area = (p * (p - self.a) * (p - self.b) * (p - self.c)) ** 0.5
      return round(area, 2)
    else:
      return "invalid"

x, y, z = map(int, input("Nhập 3 cạnh của tam giác: ").split())
t = Triangle(x, y, z)
print(t.area())
```
